```python
data_set.fillna('',inplace=True)
```

Preprocessing:


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
featuers =data_set['genres'] + ' ' + data_set['keywords'] + data_set['tagline'] + data_set['cast'] +data_set['director']
```

```python
print(featuers)
```

vectoriziation 

```python
vectorizer = TfidfVectorizer()
```

```python
vectorized_features = vectorizer.fit_transform(featuers)
```

find the similarity between movies

```python
similarity = cosine_similarity(vectorized_features)
```

```python
print(similarity)
```

```python
# create list of all movie names

movie_list = data_set['title'].tolist()
print(movie_list)
```

Recommend a movie for the user:


```python
movie_name = 'men in black'
```

```python
# find the nameclosest to the name provided by the user

closest_match = difflib.get_close_matches(movie_name,movie_list)
print(closest_match)
```

```python
#find the index of the movie that matches the titlein the dataframe

movie_index = data_set[data_set['title'] == closest_match[0]]['index'].values[0]
print(movie_index)
```

```python
# getting a similarity list of all the movies based on the index of the movie
similarity_score = list(enumerate(similarity[movie_index]))
```

```python
# Sort the movies based on similarity 
sorted_list = sorted(similarity_score, key= lambda x:x[1], reverse=True)
print(sorted_list)
```

```python
# print the name of the top 5 movies
print('Top 5 movies suggested for you : \n')

for movie ,i in zip(sorted_list[1:6],range(1,6)): 
    print(i, '.' , data_set.iloc[movie[0]].title)

```

Recommindation system:

```python
def recommend(movie):
    
    closest_match = difflib.get_close_matches(movie,movie_list)
    movie_index = data_set[data_set['title'] == closest_match[0]]['index'].values[0]
    similarity_score = list(enumerate(similarity[movie_index]))
    sorted_list = sorted(similarity_score, key= lambda x:x[1], reverse=True)
    print('Top 5 movies suggested for you : \n')
    for movie ,i in zip(sorted_list[1:6],range(1,6)):

        print(i, '.' , data_set.iloc[movie[0]].title)
```

```python
movie = input('Enter movie name : ')

recommend(movie)
```
